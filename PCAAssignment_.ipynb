```python
def capping(data, columns):
    for i in columns:
        Q1 = data[i].quantile(0.25)
        Q3 = data[i].quantile(0.75)
        IQR = Q3 - Q1
        upper_bound = Q3 + (1.5 * IQR)
        lower_bound = Q1 - (1.5 * IQR)
        data[i] = data[i].clip(lower = lower_bound, upper = upper_bound)
```

```python
capping(data, ['Malic','Ash','Alcalinity','Magnesium','Proanthocyanins','Color','Hue'])
```

### K-Means without PCA

```python
wcss = []
k = range(1,11)
for i in k:
    kmeans = KMeans(n_clusters=i, random_state = 42)
    kmeans.fit(data.drop('Type', axis = 1))
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(k, wcss, marker = 'o', linestyle = '-', color = 'blue')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
data['Cluster'] = kmeans.fit_predict(data.drop('Type',axis = 1))
```

```python
score = silhouette_score(data.drop(columns=['Type','Cluster'],axis = 1), data['Cluster'])
```

```python
print(f'Silhouette Score: {score}')
```

### K-Means with PCA

```python
wcss1 = []
k = range(1,11)
for i in k:
    kmeans = KMeans(n_clusters=i, random_state = 42)
    kmeans.fit(cl_pca)
    wcss1.append(kmeans.inertia_)
```

```python
plt.plot(k, wcss1, marker = 'o', linestyle = '-', color = 'blue')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
```

```python
kmeans1 = KMeans(n_clusters = 3, random_state = 42)
cl_pca['Cluster'] = kmeans1.fit_predict(cl_pca)
```

```python
score1 = silhouette_score(cl_pca.drop(columns=['Cluster'],axis = 1), cl_pca['Cluster'])
```

```python
print(f'Silhouette Score: {score1}')
```

Silhouette Score improved little bit after applying pca.
The Elbow Method indicated 3 optimal clusters, confirmed by the silhouette score of {score1:.3f}, suggesting a well-defined cluster structure. The data points are reasonably well-separated and cohesive within their respective clusters, indicating that KMeans with k=3 provides a good balance between compactness and separation.

Practical Implications of Using PCA and Clustering:
PCA simplifies high-dimensional datasets by transforming correlated features into fewer uncorrelated components, making visualization and computation more efficient without significant loss of information.

Clustering (K-Means) helps uncover hidden patterns or natural groupings within the data, which can support decision-making, customer segmentation, or anomaly detection.
Together, these techniques allow analysts to identify structure in complex data and improve model interpretability and performance.

Recommendations:Use PCA when your dataset has many correlated features or suffers from high dimensionality, to reduce redundancy and improve efficiency.

Use K-Means clustering when you need to discover underlying group structures or classify unlabeled data into meaningful categories.

Combining both is ideal when you want to visualize and cluster high-dimensional data, as PCA enhances the effectiveness and clarity of clustering results.
